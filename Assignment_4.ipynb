**Square Matrices**

Are matrices that have the same row and column sizes. We could say a matrix is square if  **i=j** .

```python
def mat_desc(matrix):
    is_square = True if matrix.shape[0] == matrix.shape[1] else False 
    print(f'Matrix:\n{matrix}\n\nShape:\t{matrix.shape}\nRank:\t{matrix.ndim}\nIs Square: {is_square}\n')
```

```python
square_mat = np.array([
      [2,3,7],
      [1,5,4],              
      [4,1,6] 
])
mat_desc(square_mat)
```

**Null Matrix**

Is a matrix that has no elements. It is always a subspace of any vector or matrix.

```python
def mat_desc(matrix):
    if matrix.size > 0:
        is_square = True if matrix.shape[0] == matrix.shape[1] else False 
        print(f'Matrix:\n{matrix}\n\nShape:\t{matrix.shape}\nRank:\t{matrix.ndim}\nIs Square: {is_square}\n')
    else:
        print('Matrix is Null')
```

```python
null_mat = np.array([
      [2,3,7],
      [1,5,4],              
      [4,1,6] 
])
mat_desc(null_mat)
```

**Zero Matrix**

It can be any rectangular matrix but with all elements having a value of 0.

```python
zero_mat_row = np.zeros((2,3,7))
zero_mat_sqr = np.zeros((1,5,4))
zero_mat_rct = np.zeros((4,1,6))

print(f'Zero Row Matrix: \n{zero_mat_row}')
print(f'Zero Square Matrix: \n{zero_mat_sqr}')
print(f'Zero Rectangular Matrix: \n{zero_mat_rct}')
```

**Ones Matrix**

It is just like the zero matrix, can be any rectangular matrix but all of its elements are 1s instead of 0s.

```python
ones_mat_row = np.ones((2,3,7))
ones_mat_sqr = np.ones((1,5,4))
ones_mat_rct = np.ones((4,1,6))

print(f'Ones Row Matrix: \n{ones_mat_row}')
print(f'Ones Square Matrix: \n{ones_mat_sqr}')
print(f'Ones Rectangular Matrix: \n{ones_mat_rct}')
```

**Identity Matrix**
